```python
# 'grades' facts to suggest the specialization based on the grades selected by user. 
with ruleset('grades'):
    @when_all((m.spec == 'AI'))
    def spec_ai(m):
        m.assert_fact({ 'suggestion': 'You should go with AI Specialization'})

    @when_all((m.spec == 'Data Engineering'))
    def spec_de(m):
        m.assert_fact({ 'suggestion': 'You should go with Data Engineering Specialization'})
      
    @when_all((m.spec == 'Information Security'))
    def spec_ns(m):
        m.assert_fact({ 'suggestion': 'You should go with Information Security Specialization'})

    @when_all((m.spec == 'Mobile Computing'))
    def spec_mob(m):
        m.assert_fact({ 'suggestion': 'You should go with Mobile Computing Specialization'})

    @when_all((m.spec == 'general'))
    def spec_gen(m):
        m.assert_fact({ 'suggestion': 'You should go with General MTech Course and choose courses as per your interest'})

    #Output all Suggestions
    @when_all(+m.suggestion)
    def output(cs):
        print('Specialization Suggestion: {0}'.format(cs.m.suggestion))

# 'theory_proj' facts to suggest whether user should do thesis or projects for completing the credits. 
with ruleset('theory_proj'):
    @when_all((m.theory == 'theory'))
    def thesis(e):
        e.assert_fact({ 'suggestion': 'You should do Thesis'})
    @when_all((m.practical == 'projects'))
    def project(e):
        e.assert_fact({ 'suggestion': 'You should do either Capstone or Independent Project'})

    #Output all Suggestions
    @when_all(+m.suggestion)
    def output(cs):
        print('Additional Credit Options: {0}'.format(cs.m.suggestion))

# 'clubs' fact to find the interested club for the user as per his/her choice.
with ruleset('clubs'):
    @when_all((m.activities == '1'))
    def singing(e):
        e.assert_fact({ 'suggestion': 'Join Club: Mic Drop or Audio Bytes'})
    @when_all((m.activities == '2'))
    def dancing(e):
        e.assert_fact({ 'suggestion': 'Join Club: MadToes'})
    @when_all((m.activities == '3'))
    def singing(e):
        e.assert_fact({ 'suggestion': 'Should Participate in Outdoor Sports such as Football, Volleyball etc.'})
    @when_all((m.activities == '4'))
    def singing(e):
        e.assert_fact({ 'suggestion': 'Join Club: Finnexia'})
    @when_all((m.activities == '5'))
    def singing(e):
        e.assert_fact({ 'suggestion': 'Join Club: Meraki'})
    @when_all((m.activities == '6'))
    def singing(e):
        e.assert_fact({ 'suggestion': 'Join Club: Muse'})
    @when_all((m.activities == '7'))
    def singing(e):
        e.assert_fact({ 'suggestion': 'Join Club: Machaan'})
    @when_all((m.activities == '8'))
    def singing(e):
        e.assert_fact({ 'suggestion': 'Join Club: Tasveer'})
    @when_all((m.activities == '9'))
    def singing(e):
        e.assert_fact({ 'suggestion': 'Join Club: LitSoc'})
    @when_all((m.activities == '10'))
    def singing(e):
        e.assert_fact({ 'suggestion': 'Join Club: Girl Up Udaan IIITD'})

    @when_all((m.activities == 'coding'))
    def singing(e):
        e.assert_fact({ 'suggestion': 'Join Club: Foobar or practice CP'})


    #Output all Suggestions
    @when_all(+m.suggestion)
    def output(cs):
        print('Extra-curricular Activities: {0}'.format(cs.m.suggestion))
```
